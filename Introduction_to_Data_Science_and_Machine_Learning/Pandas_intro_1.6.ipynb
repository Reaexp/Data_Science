```python
import pandas as pd
import numpy as np
```

```python
student_performance = pd.read_csv('StudentsPerformance.csv')
```

```python
student_performance.head(3)
```

```python
# Заменяем пробелы в названиях колонок на нижнее подчёркивадние
student_performance.rename(columns=lambda x:
                           x.replace(" ", "_"),
                           inplace=True)
```

```python
student_performance.head()
```

```python
student_performance.groupby('gender')
```

```python
# Среднее значение переменных для групирующей переменной
student_performance.groupby('gender').mean()
```

```python

student_performance.groupby('gender').aggregate({'math_score': 'mean'})
```

```python
student_performance.groupby('gender').aggregate({'math_score': 'mean', 'reading_score': 'mean'})
```

```python
student_performance.groupby('gender').aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score': 'mean'})
```

```python
student_performance.groupby('gender', as_index=False)\
.aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score': 'mean'})\
.rename(columns = {'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score', 'writing_score': 'mean_writing_score'})
```

```python
# групировка еще по race/ethnicity
student_performance.groupby(['gender', 'race/ethnicity'], as_index=False)\
.aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score': 'mean'})\
.rename(columns = {'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score', 'writing_score': 'mean_writing_score'})
```

```python
mean_score = student_performance.groupby(['gender', 'race/ethnicity'], as_index=False)\
.aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score': 'mean'})\
.rename(columns = {'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score',\
                   'writing_score': 'mean_writing_score'})
```

**Отберем по 5 ТОП студентов по математике внутри каждого значения по переменной 'gender'


```python
# Сортируем данные от меньшего к большему
student_performance.sort_values(['gender', 'math_score'])
```

```python
# Сортируем данные в другом порядке
student_performance.sort_values(['gender', 'math_score'], ascending=False)
```

```python
# Групперуем дата фрейм по переменной пол
student_performance.sort_values(['gender', 'math_score'], ascending=False)\
.groupby('gender').head(5)
# Получаем топ 5 математиков юношей и топ 5 математиков девушек
```

Как создавать новые колонки датафреймов

```python
# 1 способ
student_performance['total_score'] = student_performance.math_score + student_performance.reading_score\
+ student_performance.writing_score 
```

```python
student_performance.head()
```

```python
# 2 способ
student_performance = student_performance.assign(total_score_log = np.log(student_performance.total_score))
```
